```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
Tc_100 =2.275
Tc_80 = 2.28
Tc_60 = 2.285
Tc_40 = 2.295
```

```python
x = [40,60, 80, 100]
y = [Tc_40*40,Tc_60*60,Tc_80*80, Tc_100*100]
```

```python
linregress(x,y)
```

```python
Tc_inf_m = linregress(x,y).slope
a = linregress(x,y).intercept
```

```python
fig = plt.figure(figsize=(6,3.7))
ax = fig.gca()

ax.scatter(x,y, label=r'$Tc(L) \times L$', marker='x', color='teal')
ax.plot(np.arange(0,110),[Tc_inf_m*i for i in range(110)], label=r'$Tc_{inf} \times L + a$',linestyle='dotted', color= 'salmon')
ax.legend()

ax.set_xlabel('L')
ax.set_ylabel(r'$L \times T_c$')

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 14}
plt.rc('font', **font)

ax.set_facecolor((0.98, 0.98, 0.98))
```

```python

#
Tc_inf = 2/np.log(1+ np.sqrt(2))
rel_error = abs(Tc_inf_m- Tc_inf)/Tc_inf
rel_error*100
```
